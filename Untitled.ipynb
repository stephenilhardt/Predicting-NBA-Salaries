```python
import pandas as pd
import numpy as np

import rookies
import salaries

import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format = 'svg'
```

```python
salariesurl = 'http://www.espn.com/nba/salaries/_/year/{}/page/{}'
rookiesurl = "https://www.basketball-reference.com/play-index/draft_finder.cgi?request=1&year_min=2005&year_max=2015&round_min=1&round_max=2&college_id=0&pick_overall_min=1&pick_overall_max=60&pos_is_g=Y&pos_is_gf=Y&pos_is_f=Y&pos_is_fg=Y&pos_is_fc=Y&pos_is_c=Y&pos_is_cf=Y&order_by=year_id"

```

```python
sal_df = salaries.get_salary_data(salariesurl)
```

```python
rook_df = rookies.get_rookie_data(rookiesurl)
```

```python
sal_df.columns = [sal_df.iloc[0,0], sal_df.iloc[0,1], sal_df.iloc[0,2], sal_df.iloc[0,3], 'YEAR']
```

```python
rook_df.head()
```

```python
sal_df.dropna(inplace=True)
sal_df.drop_duplicates(inplace=True)
```

```python
sal_df.drop(0,axis=0, inplace=True)
```

```python
sal_df['NAME'] = [name.split(',') for name in sal_df['NAME']]
sal_df['NAME'] = [name[0] for name in sal_df['NAME']]
```

```python
sal_df['NAME'] = [name[0] for name in sal_df['NAME']]
```

```python
rook_df.head()
```

```python
rook_df.dropna(inplace=True)
rook_df.drop_duplicates(inplace=True)
```

```python
rook_df.head()
```

```python
rook_df.drop(0, axis = 0, inplace=True)
```

```python
rook_df.head()
```

```python
rook_df['NAME'] = rook_df['player']
```

```python
rook_df.head()
```

```python
df = pd.merge(rook_df, sal_df, on='NAME', how='outer')
```

```python
df.head()
```

```python
df['year_id'] = df['year_id'].astype(float)
```

```python
df['NBA_YEARS'] = df['YEAR'] - df['year_id']
```

```python
df.head()
```

```python
df['SALARY'] = [str(sal) for sal in df['SALARY']]
df['SALARY'] = [sal.replace('$','') for sal in df['SALARY']]
df['SALARY'] = [sal.replace(',','') for sal in df['SALARY']]
df.dropna(inplace=True)
df['SALARY'] = pd.to_numeric(df['SALARY'])
```

```python
nbayears = df.groupby('NBA_YEARS').mean()
```

```python
nbayears.drop(['year_id', 'YEAR'], axis=1, inplace=True)
nbayears.reset_index(inplace=True)
```

```python
nbayears.drop([0,1,2,3], inplace=True)
nbayears
```

```python


sns.lineplot(x = 'NBA_YEARS', y='SALARY', data = nbayears)
plt.title('Salary By Years In NBA')
plt.xlabel('Years In NBA')
plt.ylabel('Average Salary In Millions')
plt.savefig('salaries.svg', format='svg')
```
