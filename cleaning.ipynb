```python
adv_percentages = ['per', 'blk_pct', 'vorp', 'trb_pct','fg3a_per_fga_pct', 'obpm','ws_per_48', 'bpm','drb_pct','stl_pct',
                  'tov_pct','ts_pct','usg_pct','dbpm','fta_per_fga_pct','ast_pct','orb_pct','mp']

for perc in adv_percentages:
    advanced[perc] = [pd.to_numeric(p) for p in advanced[perc]]

```

```python
adv_ints = ['mp', 'g', 'age', 'dws', 'ws', 'ows','year']

for stat in adv_ints:
    advanced[stat] = [pd.to_numeric(s) for s in advanced[stat]]
```

```python
advanced_perc = advanced[['year','player', 'per', 'blk_pct', 'vorp', 'trb_pct','fg3a_per_fga_pct', 'obpm','ws_per_48', 'bpm','drb_pct','stl_pct',
                  'tov_pct','ts_pct','usg_pct','dbpm','fta_per_fga_pct','ast_pct','orb_pct']]

advanced_perc
```

```python
advanced_stat = advanced[['year', 'player', 'g', 'dws', 'ws', 'ows','mp']]

advanced_stat
```

```python
grouped_advanced_perc = advanced_perc.groupby(['year', 'player']).mean()
grouped_advanced_stat = advanced_stat.groupby(['year', 'player']).sum()
```

```python
grouped_advanced_perc.reset_index(inplace=True)
grouped_advanced_stat.reset_index(inplace=True)
```

```python
grouped_advanced_perc
```

```python
ints = ['age','ast','blk','drb','fg','fg2','fg2a','fg3','fg3a','fga','ft','fta','g','gs','mp','orb','pf','pts','stl','tov','trb']

for stat in ints:
    stats[stat] = stats[stat].astype(int)
```
